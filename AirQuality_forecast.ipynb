## 데이터 정제 
- 'pm2_5'컬럼에 결측치가 있으면 데이터 드롭
- 'pm2_5'컬럼의 값에 따라 WHO 기준 '예보 등급' 결정
  - 76 이상 : '매우 나쁨'
  - 36 이상 76 미만 : '나쁨'
  - 16 이상 36 미만 : '보통'
  - 16 미만 : '좋음'

```python
# 'pm2_5' 컬럼을 기준으로 결측치 존재할 경우 드롭
df_drop = df.dropna(subset=['pm2_5'])
```

```python
print(len(df_drop))
df_drop.head()
```

```python
def get_level(pm2_5):
    if pm2_5 >= 76:
        return '매우 나쁨'
    elif pm2_5 >= 36 and pm2_5 < 76:
        return '나쁨'
    elif pm2_5 >= 16 and pm2_5 < 36:
        return '보통'
    elif pm2_5 < 16 :
        return '좋음'
    
df_drop['예보 등급'] = df_drop.loc[:, ('pm2_5')].apply(get_level)
```

```python
df_drop.head()
```

## 데이터 카운트

```python
forecast_counts = df_drop['예보 등급'].value_counts()
forecast_counts
```

```python
forecast_df = forecast_counts.reset_index()
forecast_df
```

```python
forecast_df.info()
```

## 카테고리화 및 정렬 기준 정의
### pd.Categorical(values, categories, ordered)
- **values** : list-like 
  - The values of categorical.
- **categories** : Index-like(unique)
  - The unique categories for this categorical. 
- **ordered** : bool
  - Whether or not this categorical is treated as a ordered categorical.
  - If True, the resulting categorical will be orderd.
  - An ordered categorical respects, when sorted, the order of its 'categories' attribute.

```python
ordered_category = ['매우 나쁨', '나쁨', '보통', '좋음']
forecast_df['예보 등급'] = pd.Categorical(forecast_df['예보 등급'], categories=['매우 나쁨', '나쁨', '보통', '좋음'], ordered=True)
```

```python
forecast_df = forecast_df.sort_values(['예보 등급'])
```
